```python
import pandas as pd
reviews=pd.read_csv('drive/MyDrive/DataSet/winemag-data-130k-v2.csv')
reviews.head(15)
```

```python
#Cuenta cuantas veces aparece cada puntuación
reviews.groupby('points').points.count()
```

```python
#Para obtener el vino mas barato de cada puntuación
reviews.groupby('points').price.min()
```

```python
#Ordena los vinos segun la primera bodega donde fueron guardados
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
#El mejor vino por pais y provincia
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
#Genera un cuadro de estadisticas basicas ordenado por pais imprimir el precio minimo y el maximo
reviews.groupby(['country']).price.agg([len, min, max])
```

```python
#Usando indices multiples
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
#devolver la variable a indice normal
mi = countries_reviewed.index
type(mi)
countries_reviewed.reset_index()
```

```python
countries_reviewed.sort_values(by='len')
```

```python
#Ordenando por orden descendente
countries_reviewed.sort_values(by='len', ascending=False)
```

```python
#ordenando por el valor del indice
countries_reviewed.sort_index()
```

```python
#Ordenando por mas de una columna al mismo tiempo
countries_reviewed.sort_values(by=['country', 'len'])
```

```python
#El vino de mejor calidad con el precio mas barato
reviews.groupby('price').points.max()
```

```python
#los ordena por variedad y saca el precio minimo y maximo que existe de cada tipo
price_extremes=reviews.groupby(['variety']).price.agg([min, max])
price_extremes
```

```python
sorted_varieties = price_extremes.sort_values(["min","max"], ascending=[False,True])
sorted_varieties
```

```python
#Agrupados por orden de catador dependiento de la media de puntos que ha otorgado
reviews.groupby("taster_name").points.mean()
```

```python
reviews.groupby(['country', 'variety']).variety.count().sort_values(ascending=False)
```
